```python
# Defining the function 
def null_detail(df):
  func_df= pd.DataFrame(index= df.columns, columns=["datatype", "total_values", "na_values", "non_na_values", "%na_values", "unique_count"])
  func_df["datatype"]= [type(i) for i in list(df.columns)]
  func_df["total_values"]= [len(df[i]) for i in list(df.columns)]
  func_df["na_values"]= [df[i].isnull().sum() for i in list(df.columns)]
  func_df["non_na_values"]= [(~df[i].isnull()).sum() for i in list(df.columns)]
  func_df["%na_values"]= [df[i].isnull().mean()*100 for i in list(df.columns)]
  func_df["unique_count"]= [len(df[i].unique()) for i in list(df.columns)]
  return func_df
```

##**7. Which hotel has a higher rate of returning customers?**

```python
# table for hotelwise total guests
total_guests = df_hotel.groupby(["hotel"])["is_repeated_guest"].count().reset_index().rename(columns = {"hotel" : "Hotel Type", "is_repeated_guest" : "Total Families"})

# table for hotelwise repeated guests.
df1 = df_hotel[df_hotel["is_repeated_guest"] == 1]
df_is_repeated_guest = df1.groupby(["hotel"])["is_repeated_guest"].count().reset_index().rename(columns = {"hotel" : "Hotel Type", "is_repeated_guest" : "Repeated Families"})

# Merging above tables.
repeated_guest = pd.merge(total_guests, df_is_repeated_guest, how = "left", left_on = "Hotel Type", right_on = "Hotel Type")
repeated_guest

# creating new columns for percentage of repeated families.
repeated_guest["Percentage repeated families"] = round((repeated_guest["Repeated Families"] * 100) / repeated_guest["Total Families"], 2)
repeated_guest
```

```python
#graph of the repeated families table .
plt.title("customer loyalty")
sns.barplot(x = "Hotel Type", y = 'Percentage repeated families', data =repeated_guest)
```

From the above graph it is clear that highest rate of returning customers are from the resort hotel.

 ## **16. Which hotel has the highest cancellation rate, the city or the resort?** 

```python
# table for total bookings hotelwise.
df_total_bookings = df_hotel.groupby(["hotel"])["is_canceled"].count().reset_index().rename(columns ={"hotel" : "Hotel Type", "is_canceled" : "Total Bookings"})

#table for cancelled bookings hotelwise.
df_cancelled_booking = df_hotel[df_hotel["is_canceled"]==1]
df_cancelled_booking = df_cancelled_booking.groupby(["hotel"])["is_canceled"].count().reset_index().rename(columns = {"hotel" : "Hotel Type", "is_canceled" : "Canceled Bookings"})

# Joining table for hotelwise total bookings and cancelled bookings.
df_hotel_booking_cancellation = pd.merge(df_total_bookings, df_cancelled_booking, how = "left", left_on = "Hotel Type", right_on ="Hotel Type")

# Adding a column of percentage cancelled booking.
df_hotel_booking_cancellation["Percentage Cancellation"] = round((df_hotel_booking_cancellation["Canceled Bookings"] * 100) / df_hotel_booking_cancellation["Total Bookings"], 2)
df_hotel_booking_cancellation
```

```python
# Plotting the graph of the hotel booking cancellation table in order to drive the insights.
plt.title("Hotel Booking cancellation rate")
sns.barplot(x = "Hotel Type", y = 'Percentage Cancellation', data = df_hotel_booking_cancellation)
# plt.figure["figure.figsize] = (10,5)
```

1. About 30% of hotel reservations for city hotels and 24% for resort hotels are cancelled.
